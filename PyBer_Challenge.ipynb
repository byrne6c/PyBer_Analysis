### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

### Create DataFrames for Each Type of City

```python
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_per_type = pyber_data_df.groupby(["type"]).count()["ride_id"]

total_rides_rural = total_rides_per_type["Rural"]
total_rides_suburban = total_rides_per_type["Suburban"]
total_rides_urban = total_rides_per_type["Urban"]

print(total_rides_per_type)
```

```python
# 2. Get the total driver count for each city type

total_drivers_per_type = pyber_data_df.groupby(["type"]).sum()["driver_count"]

total_drivers_rural = total_drivers_per_type["Rural"]
total_drivers_suburban = total_drivers_per_type["Suburban"]
total_drivers_urban = total_drivers_per_type["Urban"]

print(total_drivers_per_type)
```

```python
#  3. Get the total amount of fares for each city type

total_fares_per_type = pyber_data_df.groupby(["type"]).sum()["fare"]

total_fares_rural = total_fares_per_type["Rural"]
total_fares_suburban = total_fares_per_type["Suburban"]
total_fares_urban = total_fares_per_type["Urban"]

print(total_fares_per_type)
```

```python
#  4. Get the average fare per ride for each city type. 

avg_urban_fare_per_ride = (urban_cities_df["fare"].sum() ) / (urban_cities_df["ride_id"].count())
avg_suburban_fare_per_ride = (suburban_cities_df["fare"].sum() ) / (suburban_cities_df["ride_id"].count())
avg_rural_fare_per_ride = (rural_cities_df["fare"].sum() ) / (rural_cities_df["ride_id"].count())

print(avg_urban_fare_per_ride)
print(avg_suburban_fare_per_ride)
print(avg_rural_fare_per_ride)
```

```python
# 5. Get the average fare per driver for each city type. 

avg_urban_fare_per_driver = (urban_cities_df["fare"].sum() ) / (urban_cities_df["driver_count"].sum())
avg_suburban_fare_per_driver = (suburban_cities_df["fare"].sum() ) / (suburban_cities_df["driver_count"].sum())
avg_rural_fare_per_driver = (rural_cities_df["fare"].sum() ) / (rural_cities_df["driver_count"].sum())
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary = { 'type': ['Rural' , 'Suburban' , 'Rural'] ,
    'Total Rides' : [total_rides_rural , total_rides_suburban , total_rides_urban] , 
    'Total Drivers' : [total_drivers_rural, total_drivers_suburban , total_drivers_urban] ,
    'Total Fares' : [total_fares_rural , total_fares_suburban , total_fares_urban] ,
    'Average Fare per Ride' : [ avg_rural_fare_per_ride , avg_suburban_fare_per_ride , avg_urban_fare_per_ride ], 
    'Average Fare per Driver' : [ avg_rural_fare_per_driver, avg_suburban_fare_per_driver, avg_urban_fare_per_driver]
                }


pyber_summary_df = pd.DataFrame(pyber_summary)
pyber_summary_df.style
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

pyber_fare_data = pyber_data_df.groupby(["type", "date"]).sum()["fare"]

pyber_fare_data_df = pd.DataFrame(pyber_fare_data)

pyber_fare_data_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.

pyber_fare_data_df = pyber_fare_data_df.reset_index()
pyber_fare_data_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

total_fare_by_date = pyber_fare_data_df.pivot(index = "date", columns = "type" , values = 'fare')
total_fare_by_date
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

total_fare_by_date_df = total_fare_by_date.loc['2019-01-01' : '2019-04-28']

total_fare_by_date_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
total_fare_by_date_df.index = pd.to_datetime(total_fare_by_date_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
total_fare_by_date_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.

weekly_fare_data = total_fare_by_date_df.resample("W").sum()
weekly_fare_data
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

weekly_fare_data.plot(figsize = (20 , 10) )

# Create a title & labels.
plt.title("Total Fare by City Type")
plt.ylabel("Fare ($USD)")
# Set the y limit between 0 and 45.
plt.ylim(0, 2500)

plt.yticks(np.arange(0, 2500, step=500))
# create legend
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

plt.savefig("analysis/PyBer_fare_summary.png")
```
